## [Problem \#1](https://projecteuler.net/problem=1): Multiples of 3 and 5
If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.

```
#function for sum of all multiples of 3 and 5
def mult_3_5(number):

	mult3_5 = 0
	hold = number

  #loop through the argument (in our case: 1000)
	while number > 0 :

    #filter multiples of 3 and 5
		if number % 3 == 0 or number % 5 == 0 :
      #sum all multiples of 3 and 5
			mult3_5 = mult3_5 + number

		else:
      #exclude non-multiples of 3 and 5
			mult3_5 = mult3_5

		number = number - 1
  #return the sum of all multiples
	return (mult3_5 - hold)
```
